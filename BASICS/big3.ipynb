```python
import sklearn.metrics as metric
import sklearn.model_selection as ms
import sklearn.datasets as dt
import sklearn.neighbors as neibour
import sklearn.tree as tree
import sklearn.naive_bayes as NB
```

```python
iris = dt.load_iris()

x=iris.data
y=iris.target

xtrain,xtest,ytrain,ytest = ms.train_test_split(x,y,test_size=.3,random_state=42)
```

```python
#knn
knn = neibour.KNeighborsClassifier(n_neighbors=3)
knn.fit(xtrain,ytrain)

pred = knn.predict(xtest)
acc = metric.accuracy_score(ytest,pred)
print(acc)
var = [[.2,.2,.2,.2]]
res=knn.predict(var)[0]

iris.target_names[res]
```

```python
#decisiontree
dt = tree.DecisionTreeClassifier()
dt.fit(xtrain,ytrain)

pred = dt.predict(xtest)
acc = metric.accuracy_score(ytest,pred)
print(acc)
var = [[.2,.2,.2,.2]]
res=dt.predict(var)[0]

iris.target_names[res]


import matplotlib.pyplot as plt

plt.figure(figsize=(18,15))
tree.plot_tree(dt,filled=True)
plt.show()

```

```python
#naivebyers
n = NB.GaussianNB()
n.fit(xtrain,ytrain)

pred = n.predict(xtest)
acc = metric.accuracy_score(ytest,pred)
print(acc)
var = [[.2,.2,.2,.2]]
res=n.predict(var)[0]

iris.target_names[res]
```
