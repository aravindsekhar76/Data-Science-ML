```python
Women = [22, 30, 35, 35, 26]
Men = [25, 32, 30, 35, 29]

n=len(Men)
index=np.arange(n)
width=0.25

plt.bar(index,Women,width=width,color="r",edgecolor="black",label="Men")
plt.bar(index+width,Men,width=width,color="g",edgecolor="black",label="Women")
plt.legend()
plt.xlabel("marks")
plt.ylabel("range")
plt.title("scores")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'medal.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

countries = data['Country']
gold_medals = data['Gold_Medals']

# Create a pie chart
plt.pie(gold_medals, labels=countries)
plt.title('Gold Medal Achievements - Top 5 Countries (2016 Summer Olympics)')
plt.show()
```
