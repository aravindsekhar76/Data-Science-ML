```python
#read a csv file and convert into dataframe and display it

data={
    'year':np.random.randint(2010,2020,10),
    'country':np.random.choice(['india','usa','uk','china'],10),
    'medals':np.random.choice(['gold','silver','bronze'],10)
}


#convert the file to csv
df=pd.DataFrame(data)
df.to_csv('olympics.csv',index=False)
df.to_csv('olympics.csv')


#read the csv file
df=pd.read_csv('olympics.csv')
df
```

```python
#dataframe sorting

data={
    "country":['india (IND)','united states (USA)','great britain (GBR)','russia (RUS)','germany (GER)'],
    "gold":[46,27,26,19,17],
    "silver":[37,23,18,18,10],
    "bronze":[38,17,26,19,15]
}

df=pd.DataFrame(data)

#sort the dataframe by  medals
df.sort_values(by=['gold','silver'],ascending=False)
df
```

```python
#finding average in dataframe

data={
    'name':['sai','ram','lakshmi','varun','amal','hari','anoop','anu','anil'],
    'designation':['developer','tester','manager','accountant','hr','sales','analyst','designer','lead'],
    'salary':[10000,20000,30000,10000,20000,30000,10000,20000,50000]
}

df=pd.DataFrame(data)
# df.sort_values(by=['salary'],ascending=False)
df.groupby('designation')['salary'].mean()

```

```python
#replace nan values with 0

data=np.array([1,2,3,np.nan,12,17,np.nan,25,30])

df=pd.DataFrame(data)
df.fillna(0)
```

```python
#identify negative and positive values in a dataframe 

data={
    "company":['google','facebook','apple','microsoft','tesla'],
    "profit":[750000,500000,-300000,-200000,100000],
}

df=pd.DataFrame(data)

df['profit']=df['profit']>0
df
```
