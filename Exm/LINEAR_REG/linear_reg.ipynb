```python
# linear regression


import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data
X =  2 * np.random.rand(100,1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Create a Linear Regression model
clf = LinearRegression()

# Train the model
clf.fit(X, y)

# Make predictions
X_new = np.array([[0], [2]])
y_pred = clf.predict(X_new)

# Plot the data and the linear regression line
plt.scatter(X, y, alpha=0.5,c='g', label='Data' )
plt.plot(X_new,y_pred, color='red', label='linear regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title(' Linear Regression')
plt.show()
```

```python
# linear LinearRegression svm


import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR

# Generate  data
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Create an SVR model
clf = SVR(kernel='linear', C=1.0)

# Train the model
clf.fit(X, y.ravel())  # SVR expects 1D target array, so we use ravel() to flatten y

# Make predictions
X_new = np.array([[0], [2]])
y_pred = clf.predict(X_new)

# Plot the data and the SVM regression line
plt.scatter(X, y, alpha=0.5, label='Synthetic Data')
plt.plot(X_new, y_pred, color='red', label='SVM Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Support Vector Machine')
plt.show()
```
